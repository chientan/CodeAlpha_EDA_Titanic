Before starting the Exploratory Data Analysis, the following questions were framed
to understand the dataset clearly:

1. What is the total number of passengers in the dataset?
2. How many passengers survived and how many did not survive?
3. Does gender have an impact on passenger survival?
4. Does passenger class (Pclass) affect survival rate?
5. Is age related to survival chances?
6. Which passenger class had the highest survival rate?

```python
df.isnull().sum()
```

```python
df['Age'].fillna(df['Age'].mean(), inplace=True)
```

```python
sns.countplot(x='Sex', hue='Survived', data=df)
plt.show()
```

```python
sns.countplot(x='Pclass', hue='Survived', data=df)
plt.show()
```

```python
plt.hist(df['Age'], bins=30)
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

```python
df.groupby('Sex')['Survived'].mean()
```

Conclusion:

Females survival rate is higher
